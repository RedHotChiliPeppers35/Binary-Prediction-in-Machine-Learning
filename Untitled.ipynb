# Seperating data and labels

```python
X = data_with_Y_values.drop(columns="Y",axis=1)
Y = data_with_Y_values["Y"]
```

# Training and Test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1, stratify= Y, random_state=1)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

# Model Training --> Logistic Regression 

```python
model = LogisticRegression()
```

```python
#training the LR model with training data
```

```python
model.fit(X_train,Y_train )
```

# Model Evaluation 

```python
#Accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print("Accuracy on train data:", training_data_accuracy)
```

```python
#Accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print("Accuracy on tesy data:", test_data_accuracy)
```
